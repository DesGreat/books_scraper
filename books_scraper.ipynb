```python
## Full Scraper Build: Multi-Page Extraction and CSV Export
## Below is the complete scraping logic broken into steps for clarity
```

```python
# Import Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define Base URL and Data Containers
BASE_URL = "https://books.toscrape.com/catalogue/page-{}.html"

titles = []
prices = []
ratings = []
```

```python
# Loop Through Pages and Fetch HTML
for page in range(1, 51):
    print(f"Scraping page {page}...")
    url = BASE_URL.format(page)
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to retrieve page {page}")
        continue
    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.select('.product_pod')
```

```python
# Extract Title, Price, and Rating
for book in books:
    title = book.h3.a['title']
    titles.append(title)

    price = book.select_one('.price_color').text
    prices.append(price)

    rating_class = book.p['class'] [1]
    ratings.append(rating_class)
```

```python
# Create DataFrame and Save to CSV
df = pd.DataFrame({
     'Title': titles,
     'Price': prices,
     'Rating': ratings
 })

df.to_csv('books_data.csv', index=False)
print("Scraping complete. Data saved to books_data.csv.")
```

```python

# Convert Ratings to Numeric Values

rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5

}
df['Rating'] = df['Rating'].map(rating_map)
```

```python
# Clean Price Column
# Remove the pound symbol and convert to float:

df['Price'] = df['Price'].astype(str).str.replace('Â£', '').str.strip().astype(float)
```
